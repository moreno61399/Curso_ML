# Practica Clustering. Soluci√≥n con k-means de scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
#Cargar la matriz de datos X
X = np.loadtxt('ex6K.txt',delimiter=',')
#Mostrar las dimensiones (n) y el numero de ejemplos (m)
m = X.shape[0]
n = X.shape[1]
```

```python
plt.scatter(X[:,0], X[:,1])
plt.grid(True)
plt.show()
```

En lugar de programar el algoritmo desde cero podemos utilizar la implementaci√≥n que est√° en la librer√≠a scikit-learn. La funci√≥n se llama KMeans. Los par√°metros son
n_clusters = n√∫mero de clusters a encontrar
init=  'random' o  'k-means++', que utiliza ... 
n_init = n√∫mero de veces que se ejecuta k-means. Se devuelve la mejor soluci√≥n de todas
max_iter = n√∫mero m√°ximo de iteraciones
tol = tolerancia por la cual se fija la convergencia

```python
k = 6
init = 'random'
n_init = 10
max_iter = 100
c_pred = KMeans(n_clusters=k, init = init, n_init = n_init, max_iter = max_iter).fit_predict(X)

plt.scatter(X[:,0], X[:,1],c=c_pred)
plt.show()
```

Utilizar la medida de la silueta para adivinar el n√∫mero de cl√∫steres:

```python
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)
```
